```python
import string
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from wordcloud import WordCloud
import matplotlib.pyplot as plt
nltk.download('punkt')
nltk.download('stopwords')

f = open('test', "r", encoding="utf-8")
text = f.read()
text.replace('\n', '')

# перевод в единый регистр (например, нижний)
text = text.lower()

# удаление специальных символов
spec_chars = string.punctuation + '\xa0«»\t—…'
text = "".join([ch for ch in text if ch not in spec_chars])

# токенизация текста
text_tokens = word_tokenize(text)

# удаление стоп-слов
russian_stopwords = stopwords.words("russian")
english_stopwords = stopwords.words("english")
russian_stopwords.extend(['это', 'нею'])

text_tokens = [token.strip() for token in text_tokens if token not in english_stopwords]

# перевод в формат для библиотеки
text = nltk.Text(text_tokens)

# подсчёт статистики
fdist = FreqDist(text)
count = 30
print(fdist.most_common(count))
fdist.plot(count,cumulative=False)

# визуализация в виде облака
text_raw = " ".join(text)
wordcloud = WordCloud().generate(text_raw)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
